3.Visualising the data

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming the Excel file is in the same directory as your script
file1_path = 'TotalMonthlyRainfall.csv'
file2_path = 'TotalNumberOfRoadAccidents.csv'

# Read the first CSV file into a DataFrame
df1 = pd.read_csv(file1_path)

# Read the second CSV file into another DataFrame
df2 = pd.read_csv(file2_path)

# Plotting
plt.figure(figsize=(12, 6))

# Line plot for the first DataFrame
plt.plot(df1['Year'], df1['Rainfall'], label='Total Rainfall', marker='o', linestyle='-')

# Line plot for the second DataFrame
plt.plot(df2['Year'], df2['Road Accidents'], label='Total Road Accidents', marker='o', linestyle='-')

# Set plot title and labels
plt.title('Total Rainfall and Road Accidents Over Years')
plt.xlabel('Year')
plt.ylabel('Count')

# Display a legend for the different datasets
plt.legend()

# Show the plot
plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming the CSV files are in the same directory as your script
file1_path = 'TotalNumberOfRoadAccidents.csv'
file2_path = 'TotalCasualties.csv'

# Read the first CSV file into a DataFrame
df1 = pd.read_csv(file1_path)

# Read the second CSV file into another DataFrame
df2 = pd.read_csv(file2_path)

# Plotting
plt.figure(figsize=(12, 6))

# Scatter plot for the first DataFrame (Total Road Accidents)
plt.scatter(df1['Year'], df1['Road Accidents'], label='Total Road Accidents', color='orange', marker='o')

# Scatter plot for the second DataFrame (Total Casualties)
plt.scatter(df2['Year'], df2['Casualties'], label='Total Casualties', color='brown', marker='s')

# Set plot title and labels
plt.title('Total Road Accidents VS Casualties Over Years')
plt.xlabel('Year')
plt.ylabel('Count')

# Display a legend for the different datasets
plt.legend()

# Display data labels near the points
for i, txt in enumerate(df1['Road Accidents']):
    plt.annotate(txt, (df1['Year'][i], df1['Road Accidents'][i]), textcoords="offset points", xytext=(0,5), ha='center')

for i, txt in enumerate(df2['Casualties']):
    plt.annotate(txt, (df2['Year'][i], df2['Casualties'][i]), textcoords="offset points", xytext=(0,5), ha='center')

# Show the plot
plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming the CSV file is in the same directory as your script
file_path = 'TotalCasualties.csv'

# Read the CSV file into a DataFrame
df = pd.read_csv(file_path)

# Plotting a grouped bar chart
plt.figure(figsize=(12, 6))

# Extracting columns for plotting
categories = df.columns[2:]  # Exclude 'Year' and 'Casualties' columns

# Calculate the width of each bar group
bar_width = 0.15  # Increased bar width

# Calculate the offset for each group of bars
offset = (len(categories) - 1) * bar_width / 2

# Plot each category as a separate group of bars with adjusted positions
for i, category in enumerate(categories):
    plt.bar(df['Year'] + i * bar_width - offset, df[category], width=bar_width, label=category)

# Set plot title and labels
plt.title('Distribution of Casualties by Category Over Years')
plt.xlabel('Year')
plt.ylabel('Count')

# Display a legend for the different categories
plt.legend()

# Show the plot
plt.show()










```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming the CSV file is in the same directory as your script
file_path = 'AgeDistributionOfVehicles.csv'

# Read the CSV file into a DataFrame
df = pd.read_csv(file_path)

# Filter out the 'Total' row
df_total = df[df['Age'] == 'Total']

# Extracting the vehicle distribution from the 'Total' row
vehicle_distribution = df_total[[' Cars', ' Motorcycles', ' Buses', ' Goods and Vehicles']].iloc[0]

# Plotting the pie chart for vehicle distribution
plt.figure(figsize=(8, 8))
plt.pie(vehicle_distribution, labels=vehicle_distribution.index, autopct='%1.1f%%', colors=['brown', 'green', 'orange', 'red'])
plt.title('Distribution of Vehicles')
plt.tight_layout()

# Show the plot
plt.show()
```
